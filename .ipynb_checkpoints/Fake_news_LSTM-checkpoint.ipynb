```python
import numpy as np
import pandas as pd
import nltk
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head(10)
```

```python
df=df.dropna()
```

```python
df.reset_index(inplace=True)
```

```python
df.head(10)
```

```python
x = df.drop('label',axis = 1)
```

```python
y = df['label']
```

```python
x,y
```

```python
message = df.copy()
```

```python
message.shape
```

```python
ps= PorterStemmer()
corpus=[]
```

```python
for i in range(len(message)):
    review = re.sub('[^a-zA-Z]',' ',message['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

```python
voc_size=5000
```

```python
oneht_rep =[one_hot(words,voc_size)for words in corpus]
oneht_rep
```

```python
sent_len= 20
embedded_doc = pad_sequences(oneht_rep,padding='pre',maxlen=sent_len)
embedded_doc
```

```python
embedding_vector_features = 40 #dimentions
model =Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_len))
#model.add(Dropout(0.2))
model.add(LSTM(100))
#model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
xfinal = np.array(embedded_doc)
yfinal = np.array(y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(xfinal,yfinal,test_size = 0.33,random_state = 42)
```

```python
model.fit(X_train,Y_train,epochs=10,validation_data=(X_test,Y_test),batch_size=64)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
y_pred=model.predict_classes(X_test)
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
accuracy_score(Y_test,y_pred)
```
